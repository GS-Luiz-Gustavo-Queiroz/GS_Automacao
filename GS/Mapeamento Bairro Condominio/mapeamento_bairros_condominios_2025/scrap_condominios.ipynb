Vamos utilizar o OverPass para capturar os dados de condomínios residenciais de fortaleza a partir do OpenStreetMap

```python
import requests
import numpy as np
import pandas as pd
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
query = """
[out:json][timeout:60];
area["name"="Fortaleza"]["boundary"="administrative"]->.searchArea;
(
  way["building"="apartments"](area.searchArea);
  way["building"="residential"](area.searchArea);
  way["building"="condominio"](area.searchArea);
  way["building"="residencial"](area.searchArea);
  way["building"="apartamento"](area.searchArea);
  relation["building"="apartments"](area.searchArea);
  relation["building"="residential"](area.searchArea);
  relation["building"="condominio"](area.searchArea);
  relation["building"="residencial"](area.searchArea);
  relation["building"="apartamento"](area.searchArea);
);
out center tags;
"""
```

```python
response = requests.get(overpass_url, params={'data': query})
data = response.json()
```
